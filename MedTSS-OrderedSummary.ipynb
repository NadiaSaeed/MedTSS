```python
from sentence_similarity import sentence_similarity

model=sentence_similarity(model_name='distilbert-base-uncased',embedding_type='sentence_embedding')
```

```python
#Step Ordered Summary
from operator import itemgetter
ress = []
for idx2, word2 in enumerate(result_3_flat):
 sentence_a = word2
 scorr =[]
 for idx1, word1 in enumerate(bvb):
    sentence_b = word1
    score=model.get_score(sentence_a,sentence_b,metric="cosine")
    scorr.append((idx1, score))
 res = max(scorr, key = itemgetter(1))[0]
 ress.append((idx2,res))
 #print(ress)
print(ress)
```
